# Exploratory Data Analysis on Demand for a Ride-sharing Service

As part of a mock take-home challenge, we load and analyze time series data containing user logins for a ride-sharing application. We use logins as a proxy for demand.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import calplot

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
login_times = pd.read_json('../data/logins.json').set_index('login_time')
login_times['cnt'] = 1.
login_times.info()
```

As per the instruction of the challenge, we analyze the time series data in 15 minute intervals.

```python
login_times_by_15_min = login_times.resample('15T').sum()
```

```python
login_times_by_15_min.describe()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
sns.lineplot(x='login_time', y='cnt', data=login_times_by_15_min, ax=ax)
ax.set_title('Number of logins over time, binned by 15 minute intervals')
ax.set_ylabel('Number of logins')
ax.set_xlabel('Time');
```

The range of login times is from 1970-01-01 to 1970-04-13. The maximum number of logins reaches over 70 at the start of March. At first glance, it appears that March and April yield a greater number of logins on average than January and February. 

We observe some periodicity in the number of logins across time but it's difficult to discern what type of seasonality is occurring. We can see this more clearly by visualizing the magnitude of logins by calendar dates:

```python
fig, axes = calplot.calplot(login_times_by_15_min.cnt, figsize=(15, 3), colorbar=True, suptitle='Calendar Plot of Login Instances', yearlabels=False)
```

We see a weekly pattern where logins are typically low during the start of the work week and increase on the weekends. This is aligned with common sense in that customers may be more likely to go out on the weekend than a work day and would therefore require transportation. There may exist some seasonality as well, since there is typically more demand in the months of March and April than there are in January and February. This is also intuitive since people go out more frequently during warmer weather.

```python
fig, ax = plt.subplots(figsize=(15, 5))
login_times_by_15_min.hist(ax=ax)
ax.set_title('Histogram of login instances per 15 minutes')
ax.set_xlabel('Number of logins')
ax.set_ylabel('Number of 15 minute intervals');
```

The number of logins has a right-skewed distribution, where most 15 minute intervals have a login count from 0 to 10, but there are rare login spikes, where an interval can yield 60 or 70 logins.

```python
def is_business_day(date):
    return bool(len(pd.bdate_range(date, date)))
```

```python
login_times_by_15_min['is_business_day'] = login_times_by_15_min.index.map(lambda i: is_business_day(i))
```

```python
login_times_by_15_min.groupby('is_business_day').describe()
```

```python
login_times_by_15_min.is_business_day.value_counts(normalize=True).plot.pie(autopct="%.1f%%");
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
sns.boxplot(x='is_business_day', y='cnt', data=login_times_by_15_min)
ax.set_title('Box plots of number of logins grouped by whether the day is a business day or not')
ax.set_xlabel('Is a business day')
ax.set_ylabel('Number of logins');
```

```python
login_times_by_15_min['during_business_hours'] = pd.Series(login_times_by_15_min.index.date, index=login_times_by_15_min.index).apply(pd.Timestamp).apply(pd.offsets.BusinessHour().onOffset)
```

```python
login_times_by_15_min['weekday'] = login_times_by_15_min.index.weekday
```

```python
login_times_by_15_min['month'] = login_times_by_15_min.index.month
```

```python
login_times_by_15_min['hour'] = login_times_by_15_min.index.hour
```

```python
fig, axes = plt.subplots(2, 1, figsize=(20, 15))
sns.boxplot(x='hour', y='cnt', data=login_times_by_15_min, ax=axes[0])
axes[0].set_title('Box plots of number of logins grouped by hour')
axes[0].set_xlabel('Hour')
axes[0].set_ylabel('Number of logins (15 minute interval)')

sns.lineplot(data=login_times_by_15_min.groupby('hour').cnt.mean(), ax=axes[1])
axes[1].set_title('Mean number of logins by hour')
axes[1].set_xlabel('Hour of the day')
axes[1].set_ylabel('Mean number of logins (15 minute interval)');
```

```python
weekday_hours = login_times_by_15_min.groupby(['weekday', 'hour']).cnt.mean().reset_index()

fig, ax = plt.subplots(figsize=(20, 7))
sns.lineplot(x='hour', y='cnt', hue='weekday', data=weekday_hours, ax=ax)
ax.legend(['Mon', 'Tue', 'Weds', 'Thu', 'Fri', 'Sat', 'Sun'], loc='best', ncol=7)
ax.set_title('Mean number of logins by hour, grouped by week day');
```

As we can see in this plot, darker hues indicate week days later in the week. There is high demand in the early morning hours of Saturday and Sunday. In the late evening, there is comparable demand on Thursday, Friday, and Saturday, but not much on Sunday night.
