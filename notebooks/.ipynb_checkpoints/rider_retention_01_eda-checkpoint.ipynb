# Rider Retention: Exploratory Data Analysis

In this exercise, we are interested in predicting rider retention for a ride-sharing service based on a sample dataset of users who signed up for an account in January 2014. A user is considered retained if they took a trip in the last 30 days.

# 1. Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import calplot
```

```python
data = pd.read_json('../data/ultimate_data_challenge.json')
data['signup_date'] = pd.to_datetime(data.signup_date)
data['last_trip_date'] = pd.to_datetime(data.last_trip_date)
data.info()
```

The average rating of driver, average rating by driver, and phone device are the only features that contain missing values.

# 2. Last Trip Date and Active Users

```python
counter = data.copy()
counter['cnt'] = 1.
calplot.calplot(counter.groupby('last_trip_date').cnt.count());
```

As expected, there are several users who have had a recent trip in June (in the last 30 days). However, there is a non-trivial amount of users who have had their last trip in January. We set our last active trip date to the maximum last trip date in the dataset.

```python
last_active_trip_date = data.last_trip_date.max() - pd.Timedelta(30, unit='d')
last_active_trip_date
```

```python
data.loc[data.last_trip_date >= last_active_trip_date, 'is_active'] = True
data.loc[data.last_trip_date < last_active_trip_date, 'is_active'] = False
```

```python
fig, ax = plt.subplots()
data.is_active.value_counts(normalize=True).plot.pie(autopct='%.1f%%')
ax.set_title('Active User Composition in Sample Dataset');
```

There are only about 38% of users who are considered active. We should consider resampling techniques to combat this class imbalance.

```python
active_users = data[data.is_active]
((active_users.last_trip_date - active_users.signup_date) >= pd.Timedelta(180, unit='d')).value_counts(normalize=True)
```

More than 99.5% of active users have not been on the platform for more than 6 months, according to last trip date.

```python
account_age = (data.last_trip_date - data.signup_date).dt.days
account_age = pd.concat([account_age, data.is_active], axis=1).rename(columns={0: 'account_age'})
fig, ax = plt.subplots()
sns.boxplot(x='is_active', y='account_age', data=account_age, ax=ax, showmeans=True)
ax.set_title('Box plots of account age (grouped by user activity');
```

```python
account_age.groupby('is_active').account_age.describe()
```

# 3. Sign Up Date

```python
counter = data.copy()
counter['cnt'] = 1.
calplot.calplot(counter.groupby('signup_date').cnt.count());
```

All users have signed up in January, many of which have signed up on the weekend.

# 4. City

```python
fig, ax = plt.subplots()
data.city.value_counts(normalize=True, dropna=False).plot.pie(autopct="%.1f%%")
ax.set_title('City Composition in Sample Dataset');
```

Winterfell comprises a large part of the dataset at about 47%, followed by Astapor at 33%, and King's Landing at 20%. There  are no observations with missing cities.

```python
tmp = data.groupby('is_active').city.value_counts(normalize=True).to_frame().rename(columns={'city': 'proportion'}).reset_index()
fig, ax = plt.subplots()
sns.barplot(x='city', y='proportion', hue='is_active', data=tmp, ax=ax)
ax.set_title('City proportion bar plot grouped by user activity');
```

A high proportion of active users are in King's Landing compared to the proportion of non-active users who operate in that city. A higher proportion of non-active users are in Winterfell and Astapor City. The city in which a user rides could be a discriminatory predictor in whether or not they are active.

# 9. Distance

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.histplot(x='avg_dist', hue='is_active', data=data, bins=20, ax=axes[0])
axes[0].set_xlabel('Average Distance')
axes[0].set_ylabel('Number of users')
axes[0].set_title('Histogram of the average distance (grouped by user activity)')

sns.boxplot(x='is_active', y='avg_dist', data=data, showmeans=True)
axes[1].set_title('Box plots of the average distance (grouped by user activity)')

fig.tight_layout()
```

```python
data.groupby('is_active').avg_dist.describe()
```

Possibly counter to intuition, there does not seem to be a dramatic difference between the distributions of average distance taken between active and non-active users. If anything, inactive users have a higher mean average distance (~6.2 vs ~5.1), but also a higher variance. One theory is that active users are more willing to use ride-sharing to travel short distances and as a result, are more likely to use the service for more types of travel situations.

# 10. Surge
## 10.1 Average Surge
The average surge multiplier on all the users trips

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.histplot(x='avg_surge', hue='is_active', data=data, bins=20, ax=axes[0])
axes[0].set_xlabel('Average Surge')
axes[0].set_ylabel('Number of users')
axes[0].set_title('Histogram of the average surge (grouped by user activity)')

sns.boxplot(x='is_active', y='avg_surge', data=data, showmeans=True)
axes[1].set_title('Box plots of the average surge (grouped by user activity)')

fig.tight_layout()
```

```python
data.groupby('is_active').avg_surge.describe()
```

```python
data.groupby('is_active').avg_surge.median()
```

The median surge multiplier for both groups is 1, meaning no surge pricing was applied. The distributions across groups are almost identical, with mean average both around 1.07. Non-active users have a slightly higher variance due to outliers.

## 10.2 Surge Percentage
The percentage of trips taken with a surge multiplier greater than 1

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.histplot(x='surge_pct', hue='is_active', data=data, bins=20, ax=axes[0])
axes[0].set_xlabel('Surge Percentage')
axes[0].set_ylabel('Number of users')
axes[0].set_title('Histogram of the surge percentage (grouped by user activity)')

sns.boxplot(x='is_active', y='surge_pct', data=data, showmeans=True)
axes[1].set_title('Box plots of the surge percentage (grouped by user activity)')

fig.tight_layout()
```

```python
data.groupby('is_active').surge_pct.describe()
```

There is a noticeable difference in surge percentage across groups: Active users tend to have a higher percentage of trips with surge pricing (9.15) than non-active users (8.67). Non-active users have a higher variance in percentage of surge-priced trips (22.2 vs 15.6) than active users.

```python
data.groupby('is_active').surge_pct.median()
```

Still, for both groups, the median surge percentage is 0%, so it may be an unreliable predictor of user activity.

# 13. Summary

- The dataset analyzed had 12 features (3 categorical, 2 datetime, 7 numerical).
- Using the last trip taken date feature, we derived a target variable called `is_active` that represented whether the user has taken a ride within the last 30 days. There is about a 62:38 ratio between non-active and active users, so techniques to combat this class imbalance must be used to avoid biasing our learner.
- More than 99.5% of active users have not been on the platform at least 6 months
- The average driver rating, average rating given by the driver, and the phone device contained a relatively small amount of missing values. We will impute these missing values using the median.
- Categorical features demonstrated the most obvious discrimination between active and in-active users
- Over 80% of active users own an iPhone compared to about 60% of non-active users
- About 34% of active users live in King's Landing compared to about 12% of non-active users
- About 70% of non-active users have not used the ultimate black ride option, compared to 50% of active users
- Distance and weekday percentage demonstrated the largest difference between active and non-active users
- Unsurprisingly, we found that surge multiplier and the percentage of surge-priced trips were highly correlated.
- No single pair of numerical features produced noticeable discrimination between active and non-active users.
